# Question formulation


## Question 1
**Liệu chiều cao có còn là tiêu chí tiên quyết cho từng vị trí trong bóng đá hiện đại hay không?**
* **Chi tiết:** Phân tích sự phân bố chiều cao trung bình của cầu thủ theo từng vị trí (Thủ môn, Trung vệ, Hậu vệ cánh, Tiền vệ, Tiền đạo). Kiểm định xem liệu bóng đá hiện đại (nơi kỹ thuật và tốc độ lên ngôi) có còn ưu tiên thể hình như truyền thống hay không.

### Motivation & Benefits
* **Lý do nghiên cứu:** Sự xuất hiện của những cầu thủ thấp bé nhưng thành công (ví dụ: Lisandro Martinez, Messi) đang thách thức các tiêu chuẩn cũ về thể hình.
* **Lợi ích:** Giúp phá bỏ định kiến trong tuyển chọn tài năng, tránh bỏ sót nhân tài chỉ vì thiếu chiều cao.
* **Đối tượng quan tâm:** Tuyển trạch viên (Scouts), Huấn luyện viên đội trẻ.
* **Vấn đề thực tế:** Hỗ trợ ra quyết định trong việc: Có nên loại một trung vệ tài năng chỉ vì anh ta thấp hơn tiêu chuẩn 5cm?

## Question 2
**So sánh hiệu suất thi đấu giữa 'Inverted Wingers' (Cánh nghịch chân) và 'Traditional Wingers' (Cánh thuận chân) - Xu hướng nào đang chiếm ưu thế?**
* **Chi tiết:** Phân loại cầu thủ chạy cánh dựa trên chân thuận và cánh thi đấu. So sánh số lượng bàn thắng, kiến tạo để xác định xu hướng chiến thuật hiệu quả hơn.

### Motivation & Benefits
* **Lý do nghiên cứu:** Chiến thuật bóng đá đã dịch chuyển mạnh mẽ từ tạt cánh (truyền thống) sang đột phá vào trung lộ để dứt điểm (nghịch chân).
* **Lợi ích:** Hiểu rõ về xu hướng chiến thuật và yêu cầu kỹ năng đối với cầu thủ chạy cánh hiện đại.
* **Đối tượng quan tâm:** Nhà phân tích chiến thuật, Huấn luyện viên.
* **Vấn đề thực tế:** Giúp HLV quyết định lối chơi và các lò đào tạo định hướng phát triển kỹ năng cho cầu thủ trẻ.

## Question 3
**Thời kì hoàng kim của cầu thủ thường rơi vào độ tuổi nào?**
* **Chi tiết:** Xác định độ tuổi mà tại đó hiệu suất thi đấu hoặc giá trị chuyển nhượng đạt đỉnh. Phân tích sự khác biệt về độ "chín" giữa các nhóm tuổi.

### Motivation & Benefits
* **Lý do nghiên cứu:** Hiểu vòng đời sự nghiệp là yếu tố cốt lõi trong quản lý nhân sự thể thao.
* **Lợi ích:** Tối ưu hóa chính sách chuyển nhượng và thời điểm gia hạn hợp đồng.
* **Đối tượng quan tâm:** Giám đốc thể thao, Người đại diện cầu thủ.
* **Vấn đề thực tế:** Trả lời câu hỏi: Có nên ký hợp đồng dài hạn (5 năm) với một cầu thủ đã 29 tuổi hay không?

## Question 4
**Phân tích luồng di cư bóng đá - Xu hướng cầu thủ thi đấu cho tuyển quốc gia hay nơi khác?**
* **Chi tiết:** Khám phá sự đa dạng quốc tịch trong các giải đấu. So sánh tỷ lệ cầu thủ nội địa và cầu thủ ngoại để thấy dòng chảy nhân tài ở các nước.

### Motivation & Benefits
* **Lý do nghiên cứu:** Bóng đá ngày càng toàn cầu hóa, việc nhận diện nguồn gốc nhân tài là rất quan trọng.
* **Lợi ích:** Nhận diện các "mỏ vàng" tài năng từ các khu vực như Nam Mỹ, Châu Phi,...
* **Đối tượng quan tâm:** Tuyển trạch viên quốc tế, Liên đoàn bóng đá.
* **Vấn đề thực tế:** Xác định mức độ phụ thuộc vào ngoại binh của các giải đấu lớn.

## Question 5
**Những câu lạc bộ nào là nơi cung cấp những cầu thủ "hàng rẻ, chất lượng cao"?**
* **Chi tiết:** Tìm ra các CLB sở hữu cầu thủ có chỉ số kỹ năng tốt nhưng mức lương hoặc giá trị chuyển nhượng thấp (Undervalued players).

### Motivation & Benefits
* **Lý do nghiên cứu:** Để tối ưu hóa ngân sách mà vẫn có cầu thủ chất lượng.
* **Lợi ích:** Tìm kiếm các món hời trên thị trường chuyển nhượng, giúp CLB nhỏ cạnh tranh với các ông lớn.
* **Đối tượng quan tâm:** Các CLB ngân sách hạn hẹp.
* **Vấn đề thực tế:** Giải quyết bài toán kinh tế: Xây dựng đội hình mạnh nhất với chi phí thấp nhất.

## Question 6
**Có thể dự đoán 'Giá trị cầu thủ' (Market Value) mùa tiếp theo dựa trên dữ liệu và thống kê cơ bản ở mùa gần nhất hay không?**
* **Chi tiết:** Xem xét việc áp dụng mô hình có hiệu quả không?(độ chính xác). Nếu không thì trả lời tại sao?

### Motivation & Benefits
* **Lý do nghiên cứu:** Thị trường chuyển nhượng thường bị lạm phát ảo, cần một thước đo chính xác hơn.
* **Lợi ích:** Tránh việc mua hớ hoặc bán rẻ tài năng, định giá dựa trên dữ liệu thực.
* **Đối tượng quan tâm:** Bộ phận đàm phán chuyển nhượng.
* **Vấn đề thực tế:** Cung cấp cơ sở khoa học để đàm phán lương và phí chuyển nhượng.
