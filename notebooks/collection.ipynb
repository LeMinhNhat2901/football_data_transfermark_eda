#  Data Collection

### What subject is your data about?

* **Topic & Domain:** Dữ liệu thuộc lĩnh vực **Phân tích Thể thao (Sports Analytics)**. Tập dữ liệu này cung cấp cái nhìn toàn diện về bóng đá chuyên nghiệp, bao gồm hồ sơ cầu thủ, câu lạc bộ chủ quản, vị trí, hiệu suất thi đấu và giá trị chuyển nhượng của cầu thủ.
* **Real-world context:**
    * Dữ liệu phản ánh tình hình của bóng đá toàn cầu, nơi giá trị cầu thủ được định lượng bằng tiền tệ (EUR).
    * Mô hình hóa mối quan hệ giữa phong độ cá nhân (goals, assists), đặc điểm cầu thủ (tuổi, chiều cao), và các yếu tố khác (câu lạc bộ, giải đấu) ảnh hưởng đến giá trị chuyển nhượng.
    * Dữ liệu bao gồm các giải đấu như Premier League, La Liga, Bundesliga, v.v., đại diện cho những giải đấu bóng đá hàng đầu trên thế giới.

### What is the source of your data?

* **Platform:** Kaggle Datasets.
    * **URL:** [https://www.kaggle.com/datasets/davidcariboo/player-scores](https://www.kaggle.com/datasets/davidcariboo/player-scores)
* **Original Source:** Dữ liệu gốc được trích xuất từ **Transfermarkt.com** - website có trụ sở tại Đức, được coi là một trong những cơ sở dữ liệu uy tín nhất thế giới về thị trường chuyển nhượng bóng đá.
* **Structure:** Bộ dữ liệu được tổ chức dưới dạng **Cơ sở dữ liệu quan hệ** gồm nhiều file CSV liên kết với nhau (`players.csv`, `clubs.csv`, `valuations.csv`, `games.csv`, `transfers.csv`), cho phép thực hiện các truy vấn phức tạp.
* **Collection Date:** *[03/12/2025]*
* **Last Updated by Author:** 8 month ago

### How was this data collected?
* **Collection Method:** **Automated Web Scraping**. Tác giả sử dụng các công cụ kỹ thuật (như Python) để duyệt qua cấu trúc phân cấp của trang web Transfermarkt và trích xuất thông tin theo định dạng bảng.
* **Target Population:**
    * **Phạm vi:** Hàng chục nghìn cầu thủ chuyên nghiệp từ các giải đấu lớn trên thế giới.
    * **Thời gian:** Dữ liệu kéo dài qua nhiều mùa giải (2012-2025).
* **Limitations & Biases:**
    * **Subjectivity Bias:** "Market Value" trên Transfermarkt không phải là thuật toán máy tính mà dựa trên sự thảo luận và đồng thuận của cộng đồng người dùng và chuyên gia trên diễn đàn Transfermarkt.
    * **Coverage Bias:** Dữ liệu cực kỳ chi tiết ở các giải đấu lớn nhưng có thể thiếu sót hoặc cập nhật chậm hơn ở các giải đấu nhỏ hoặc hạng thấp.

### Why did you choose this dataset?

* **Interest:** Nhóm quan tâm đến việc áp dụng khoa học dữ liệu vào lĩnh vực thể thao, đặc biệt là bài toán dự đoán giá trị chuyển nhượng của cầu thủ thông qua hiệu suất thi đấu, câu lạc bộ chủ quản... Và hơn nữa, nhóm cũng có niềm đam mê với việc chơi bóng đá.
* **Potential Questions:**
  Dữ liệu này có thể giúp nhóm trả lời các câu hỏi sau:
  * Liệu chiều cao có còn là tiêu chí tiên quyết cho từng vị trí trong bóng đá hiện đại hay không?
  * So sánh hiệu suất thi đấu giữa 'Inverted Wingers' (Cánh nghịch chân) và 'Traditional Wingers' (Cánh thuận chân) - Cái nào đang là xu thế?
  * Thời kì hoàng kim của cầu thủ thường rơi vào độ tuổi nào?
  * Phân tích luồng di cư bóng đá - xu hướng cầu thủ thi đấu cho tuyển quốc gia hay nơi khác?
  * Những câu lạc bộ nào là nơi cung cấp những cầu thủ "hàng rẻ, chất lượng cao"?
  * Có thể dự đoán 'Giá trị cầu thủ' mùa tiếp theo dựa trên dữ liệu và thống kê cơ bản ở mùa gần nhất(Tuổi, Vị trí, Bàn thắng..) hay không?
